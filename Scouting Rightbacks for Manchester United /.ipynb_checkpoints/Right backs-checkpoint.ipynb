```python
def scrape(link, cols):
    res = requests.get(link)
    soup = BeautifulSoup(res.content, 'lxml')
    tables = soup.findAll('tbody')
    rows = tables[0].findAll('tr')
    table = dict()
    
    for curElem in rows:
        if (curElem.find('th', {'scope': 'row'}) != None):
            for curr in cols:
                k = curElem.find('td',{'data-stat':curr})
                text = k.text.strip().encode()
                pretty_text = text.decode('utf-8')
                if pretty_text == '':
                    pretty_text = '0'
                if curr in table:
                    table[curr].append(pretty_text)
                else:
                    table[curr] = [pretty_text]

    players = pd.DataFrame(table)
    return players
```

```python
def plot(data, titles, names):
    fig, ax = plt.subplots(figsize=(23,15), facecolor='black')
    ax.set(facecolor = "black")

    annotations = right_backs.player.tolist()
    awb = data[data['player'] == 'Aaron Wan-Bissaka'][names].values.tolist()
    interest = data[(data[names[0]] > awb[0][0]) & (data[names[1]] >awb[0][1])]['player'].tolist()
    if 'Diogo Dalot' in interest:
        interest.remove('Diogo Dalot')
    mufc = ['Diogo Dalot','Aaron Wan-Bissaka']
    
    for i, label in enumerate(annotations):
        if label in interest:
            ax.scatter(data[names[0]].iloc[i], data[names[1]].iloc[i], color='#338BFF', s=80)
        elif label in mufc:
            ax.scatter(data[names[0]].iloc[i], data[names[1]].iloc[i], color='#FF8C22', s=80) 
        else:
            ax.scatter(data[names[0]].iloc[i], data[names[1]].iloc[i], color='purple')

    
    for i, label in enumerate(annotations):
        if label in interest or label in mufc:
            temp = label.split()
            if len(temp) == 1:
                ax.annotate(label, (data[names[0]].iloc[i], data[names[1]].iloc[i]), color='white', fontsize=14)
            elif len(temp) == 2:
                ax.annotate(temp[1], (data[names[0]].iloc[i], data[names[1]].iloc[i]), color='white', fontsize=14)
            elif len(temp) == 3:
                ax.annotate(temp[2], (data[names[0]].iloc[i], data[names[1]].iloc[i]), color='white', fontsize=14)
    
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['bottom'].set_color('white')
    ax.spines['left'].set_color('white')
    ax.tick_params(axis='x', labelsize=16, colors='white')
    ax.tick_params(axis='y', labelsize=16, colors ='white')
    
    ax.set_title(titles[0] + ' for Right-Backs in 2020-2021 European Big Five Leagues \n Criteria: Age < 31 and 90s played >= 10', fontsize=26, color='white')
    ax.set_xlabel(titles[1], fontsize=26, color='white')
    ax.set_ylabel(titles[2], fontsize=26, color='white')
    
    ax.text(7, -30, 'Author: @D_C_L_07 \n Source Data: StatsBomb through Fbref', fontsize=14, color='white')
    fig.savefig(titles[0]+'.png', bbox_inches='tight', dpi=200, facecolor='black')
    plt.show()
    return interest + mufc
```
