```python
def make_map(df):
    
    from mplsoccer.pitch import Pitch, VerticalPitch
    from mplsoccer import FontManager
    
    pitch = Pitch(pitch_type='statsbomb', pitch_color='#F0F0F0', line_color='#22312b')
    fig, ax = pitch.draw(figsize=(12, 8), constrained_layout=False, tight_layout=True)
    ax.invert_yaxis()
    fig.set_facecolor('#F0F0F0')
    
    unsuccessful_lines = pitch.lines(df[df['outcomeType/displayName'] == 'Unsuccessful'].x, df[df['outcomeType/displayName'] == 'Unsuccessful'].y,
                  df[df['outcomeType/displayName'] == 'Unsuccessful'].endX, df[df['outcomeType/displayName'] == 'Unsuccessful'].endY,
                  lw=5, transparent=True, comet=True, label='Unsuccessful',
                  color='#BA2220', ax=ax)

    successful_lines = pitch.lines(df[df['outcomeType/displayName'] == 'Successful'].x, df[df['outcomeType/displayName'] == 'Successful'].y,
                  df[df['outcomeType/displayName'] == 'Successful'].endX, df[df['outcomeType/displayName'] == 'Successful'].endY,
                  lw=5, transparent=True, comet=True, label='Successful',
                  color='#15558C', ax=ax)
    
    # types of passes

    # long passes
    pitch.scatter(df[(df['outcomeType/displayName'] == 'Successful') & (df['qualifiers/0/type/displayName'] == 'Chipped')].endX, 
              df[(df['outcomeType/displayName'] == 'Successful') & (df['qualifiers/0/type/displayName'] == 'Chipped')].endY,
              s=100, marker='D', edgecolors='black', c='white', zorder=2,
              label='long ball', ax=ax)

    # just passes
    pitch.scatter(df[~(df['qualifiers/0/type/displayName'] == 'Chipped')].endX, 
              df[~(df['qualifiers/0/type/displayName'] == 'Chipped')].endY,
              edgecolors='#22312b', c='white', s=80, zorder=2,
              label='pass', ax=ax)


    #key passes leading to shot
    pitch.scatter(df[df.isin(['KeyPass']).any(axis=1)].endX, df[df.isin(['KeyPass']).any(axis=1)].endY,
                 s=300, marker='*', edgecolors='black', c='white', zorder=2,
              label='Key Pass', ax=ax)
    
    pitch.scatter(df[df.isin(['IntentionalGoalAssist']).any(axis=1)].endX, df[df.isin(['IntentionalGoalAssist']).any(axis=1)].endY,
                 s=300, marker='*', edgecolors='black', c='yellow', zorder=2,
              label='Assists', ax=ax)

    robotto_regular = FontManager()

    # Plot the legend
    ax.legend(facecolor='#F0F0F0', edgecolor='None', fontsize=10, loc='lower left', handlelength=3)

    # Set the title
    ax.set_title('Lisandro Martinez Passes vs Leeds', fontsize=24)
    
    # subtitle
    ax.text(30, 81, "30/35 Passes Completed, 86% Success Rate", fontsize=14)
    #endnote
    ax.text(108, 2, '@D_C_L_07', fontsize=12)
    
    fig.savefig('Licha_pass_map.png', bbox_inches='tight', dpi=200, facecolor='#F0F0F0')
    
```

```python
player = game[(game['playerId'].isin([79554, 100008, 17708, 99487, 345303])) & game['period/value']==1]
```

```python
import numpy as np
```

```python
from sklearn.cluster import KMeans
```

```python
buildup_passes =player[player['type/displayName'] == 'Pass']
```

```python
buildup_passes[['x','y','endX','endY']].isna().sum()
```

```python
buildup_passes['x'] = buildup_passes['x']*1.2
buildup_passes['y'] = buildup_passes['y']*0.8
buildup_passes['endX'] = buildup_passes['endX']*1.2
buildup_passes['endY'] = buildup_passes['endY']*0.8
```

```python
buildup_passes['x'].describe()
```

```python
buildup_passes = buildup_passes[buildup_passes['x']<35]
```

```python
# Elbow Method: determining how many clusters to use in the final plots
K = range(1,10)
wss= []
for k in K:
    array = np.array(buildup_passes[['x', 'y', 'endX', 'endY']])
    kmeans = KMeans(n_clusters = k, random_state = 100)
    kmeans.fit(array)
    wss_iter = kmeans.inertia_
    wss.append(wss_iter)
    
mycenters = pd.DataFrame({'Clusters':K, 'WSS':wss})
mycenters
```

```python
plt.scatter(x=mycenters['Clusters'], y=mycenters['WSS'])
```

```python
X = np.array(buildup_passes[['x','y','endX','endY']])
kmeans = KMeans(n_clusters = 4,random_state=100)
kmeans.fit(X)
buildup_passes['cluster'] = kmeans.predict(X)
```

```python
buildup_passes.cluster.unique()
```

```python
from mplsoccer.pitch import Pitch, VerticalPitch
from mplsoccer import FontManager
```

```python
pitch = Pitch(pitch_type='statsbomb', pitch_color='#F0F0F0', line_color='#22312b')
fig, ax = pitch.draw(figsize=(12, 8), constrained_layout=False, tight_layout=True)
ax.invert_yaxis()
fig.set_facecolor('#F0F0F0')

for x in range(len(buildup_passes['cluster'])):
    
    if list(buildup_passes['cluster'])[x]==0:
        pitch.lines(xstart=list(buildup_passes['x'])[x],ystart=list(buildup_passes['y'])[x],
                    xend=list(buildup_passes['endX'])[x],yend=list(buildup_passes['endY'])[x],
                   color='#74c69d',lw=3,zorder=2,comet=True,ax=ax)
        plt.scatter(list(buildup_passes["endX"])[x],list(buildup_passes["endY"])[x],s=30, color="#74c69d", zorder=100)
        
    if list(buildup_passes['cluster'])[x]==1:
        pitch.lines(xstart=list(buildup_passes['x'])[x],ystart=list(buildup_passes['y'])[x],
                    xend=list(buildup_passes['endX'])[x],yend=list(buildup_passes['endY'])[x],
                   color='#add8e6',lw=3,zorder=2,comet=True,ax=ax)
        plt.scatter(list(buildup_passes["endX"])[x],list(buildup_passes["endY"])[x],s=30, color="#add8e6", zorder=100)
    
    if list(buildup_passes['cluster'])[x]==2:
        pitch.lines(xstart=list(buildup_passes['x'])[x],ystart=list(buildup_passes['y'])[x],
                    xend=list(buildup_passes['endX'])[x],yend=list(buildup_passes['endY'])[x],
                   color='#CBC3E3',lw=3,zorder=2,comet=True,ax=ax)
        plt.scatter(list(buildup_passes["endX"])[x],list(buildup_passes["endY"])[x],s=30, color="#CBC3E3", zorder=100)
    
    if list(buildup_passes['cluster'])[x]==3:
        pitch.lines(xstart=list(buildup_passes['x'])[x],ystart=list(buildup_passes['y'])[x],
                    xend=list(buildup_passes['endX'])[x],yend=list(buildup_passes['endY'])[x],
                   color='#FAC898',lw=3,zorder=2,comet=True,ax=ax)
        plt.scatter(list(buildup_passes["endX"])[x],list(buildup_passes["endY"])[x],s=30, color="#FAC898", zorder=100)
    
    # Set the title
ax.set_title('Top 4 Clusters of Successful Passes by Manchester United Back Five in Buildup (1st Half)', fontsize=18)
    
# subtitle
ax.text(30, 81, "30/35 Passes Completed, 86% Success Rate", fontsize=14)
#endnote
ax.text(108, -3, '@D_C_L_07', fontsize=12)
    
fig.savefig('buildup_map.png', bbox_inches='tight', dpi=200, facecolor='#F0F0F0')
    
```
