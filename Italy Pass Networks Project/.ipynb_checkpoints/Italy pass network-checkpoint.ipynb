```python
def graph(df, df_a, title, sub):
    
    temp = title.split('.')
    min_transparency = 0.3
    color = np.array(to_rgba('#d3d3d3'))
    color = np.tile(color, (len(df),1))
    c_transparency = df.pass_count / df.pass_count.max()
    c_transparency = (c_transparency * (1 - min_transparency)) + min_transparency
    color[:, 3] = c_transparency
    
    MAX_LINE_WIDTH = 18
    MAX_MARKER_SIZE = 3000
    df['width'] = (df.pass_count / df.pass_count.max() *
                           MAX_LINE_WIDTH)
    df_a['marker_size'] = (df_a['count']
                                         / df_a['count'].max() * MAX_MARKER_SIZE)

    
    pitch = Pitch(pitch_type='statsbomb', pitch_color='#023020', line_color='#c7d5cc')
    fig, ax = pitch.draw(figsize=(16,12))
    fig.set_facecolor("#023020")
    ax.invert_yaxis()
    
    for index, row in df_a.iterrows():
        pitch.annotate(row.name, xy=(1.2*row.x, 0.8*row.y), c='w', va='center', ha='center', size=16, fontweight='bold',ax=ax)

    
    pass_lines = pitch.lines(1.2*df.x, .8*df.y, 1.2*df.x_end, .8*df.y_end, lw = df.width,
                            color='w', ax=ax, zorder=1, alpha=0.5)
    
    nodes = pitch.scatter(1.2*df_a.x,.8*df_a.y,
                     s = df_a.marker_size, color = '#0D63A3', edgecolors = 'black', linewidth = 2.5, alpha = 1, zorder = 1, ax=ax)
    
    if sub >=45:
        ax.set_title(temp[0] + " before First Sub"+"('" +str(sub) + ")\n" + "2020 European Championship", fontsize=18, color='w', fontweight='bold')
    elif sub < 45:
        ax.set_title(temp[0] + " adter First Sub"+"('" +str(sub) + ")\n" + "2020 European Championship", fontsize=18, color='w', fontweight='bold')
    ax.text(55,-8,"@D_C_L_07", fontstyle='italic', fontsize=12, color='w')
    
    pitch.annotate("Node Locations: Average Start of Successful Passes\nMinimum 3 passes", (58,-3), color='w', fontsize=12, ha='center', va='center', ax=ax, fontweight='bold')
    fig.savefig(title, bbox_inches='tight', dpi=200, facecolor='#023020')
   
```

```python
def process_data(df, title, jersey_nums):
    df = df[df['teamId'] == 343]
    df['passer'] = df['playerId']
    df['recipient'] = df['playerId'].shift(-1)
    
    passes = df[df['type/displayName'] == 'Pass']
    successful = passes[passes['outcomeType/displayName'] == 'Successful']
    
    subs = df[df['type/displayName'] == 'SubstitutionOff']
    subs = subs['minute']
    firstSub = subs.min()
    if firstSub >=45:
        successful = successful[successful['minute'] < firstSub]
    elif firstSub < 45:
        successful = successful[successful['minute'] > firstSub]
    
    pas = pd.to_numeric(successful['passer'], downcast='integer')
    rec = pd.to_numeric(successful['recipient'], downcast='integer')
    
    successful['passer'] = pas
    successful['recipient'] = rec
    
    average_locations = successful.groupby('passer').agg({'x': ['mean'], 'y': ['mean', 'count']})
    average_locations.columns = ['x', 'y', 'count']
    pass_between = successful.groupby(['passer', 'recipient']).id.count().reset_index()
    pass_between.rename({'id':'pass_count'},axis='columns',inplace=True)
    
    pass_between = pass_between.merge(average_locations, left_on='passer',right_index=True)
    pass_between = pass_between.merge(average_locations, left_on='recipient',right_index=True,suffixes=['', '_end'])
    pass_between = pass_between[pass_between['pass_count']>3]
    
    average_locations.index = average_locations.index.map(jersey_nums)
    pass_between['passer'] = pass_between['passer'].map(jersey_nums)
    pass_between['recipient'] = pass_between['recipient'].map(jersey_nums)
    graph(pass_between, average_locations, title, firstSub)
    return pass_between
```

```python
jersey = {9991: 3,  20664: 19,  70141: 17,  83390: 10,  92329: 4, 101596: 24, 106968: 8, 118542: 11,
            148684: 18, 262515: 21, 273539: 5}
photo_title = "Italy vs Turkey.png"
p = process_data(df_1, photo_title, jersey)
```

```python
import networkx as nx
from networkx.algorithms.centrality import closeness_centrality, betweenness_centrality
```

```python
from tabulate import tabulate
```

```python
def network_analysis(df):
    df_graph = df[['passer', 'recipient', 'count']]
    G = nx.from_pandas_edgelist(df=df_graph, source = 'passer', target='recipient', edge_attr='count', create_using=nx.DiGraph())
    #print(G.number_of_nodes())
    #print(G.number_of_edges())
    #degrees = nx.degree(G)
    
    # who had the highest number of passers 
    #print(degrees)
    #print(max(degrees,key= lambda i : i[1]))
    
    # how central a node is, in soccer context, the main source for building up play
    from collections import OrderedDict
    cc = OrderedDict(sorted(
                    closeness_centrality(G).items(),
                    key = lambda x: x[1],
                    reverse = True))
    
    # which nodes serve as a connector, higher the number the more connector the node is
    connector = nx.betweenness_centrality(G)
    
    #print(tabulate(cc))
    #print(tabulate(connector))
    central = pd.DataFrame.from_dict(dict(cc), orient='index')
    between = pd.DataFrame.from_dict(connector, orient='index')
    between = between.sort_values(by=between.columns[0], ascending=False)
    print(tabulate(central, headers=['Player', 'Closeness Centrality'], tablefmt='fancy_grid'))
    print(tabulate(between, headers=['Player', 'Betweenness Centrality'], tablefmt='fancy_grid'))
```

```python
network_analysis(p)
```

Italy used a lopsided back 3 with chielleni, bonucci, and florenzi
Spinazzola moved higher up the pitch on the left
The centrebacks were responsible for circulating the ball
Jorginho served as the linker to progress the ball centrally and dictate play
With florenzi forming a back 3, passing lanes opened for italy to progress the ball to barella on the right,
and barella would link up with berardi to create opportunities
Italy did a lot of damage through the left side as they caused an overload
